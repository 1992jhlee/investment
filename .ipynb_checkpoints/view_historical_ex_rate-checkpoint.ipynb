```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from datetime import datetime, timedelta
plt.style.use('seaborn')
```

```python
ex_rate_df = pd.DataFrame()
for file in glob.glob('data/과거환율추이*.xls'):
    df = pd.read_excel(file)
    dates = pd.Series(df.iloc[1, :].values).dropna()
    dates = dates.apply(lambda x: datetime.strptime(x, "%Y%m월"))
    dates = dates.apply(lambda x: x.strftime("%Y/%m"))
    df = df.iloc[2:3, 2:].reset_index(drop=True)
    df.columns = dates
    df = df.T
    df.columns = ['Ex. rate']
    df['Ex. rate'] = df['Ex. rate'].apply(lambda x: float(x.replace(',', '')))
    ex_rate_df = pd.concat([ex_rate_df, df])
```

```python
ex_rate_df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 5))
sns.lineplot(x=ex_rate_df.index, y='Ex. rate', data=ex_rate_df, ax=ax)
plt.title('Historical Excange Rates', fontsize=20)
plt.xticks(np.arange(0, len(ex_rate_df), step=15), ['{}'.format(ex_rate_df.index[x]) for x in np.arange(0, len(ex_rate_df), step=15)], rotation=45)
plt.hlines(y=ex_rate_df['Ex. rate'].mean(), xmax=ex_rate_df.index.max(), xmin=ex_rate_df.index.min(), color='r', label='historical mean = {:.2f}'.format(ex_rate_df['Ex. rate'].mean()))
plt.legend(fontsize=15)
plt.show()
```
