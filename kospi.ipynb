```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
from datetime import datetime, timedelta

plt.style.use('default')
```

### 수출입 통계 데이터

```python
tradeDf = pd.DataFrame()
for file_name in glob.glob('data/수출입실적*'):
    file = pd.read_excel(file_name)
    months = file.iloc[1].dropna().values
    exports = file.iloc[2].values[1:].astype('float64')
    imports = file.iloc[4].values[1:].astype('float64')
    df = pd.DataFrame({'date': months, 'export [100m$]': exports, 'import [100m$]': imports})
    df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y%m월'))
    df['date'] = df['date'].apply(lambda x: x.strftime('%Y-%m'))
    tradeDf = pd.concat([tradeDf, df]).reset_index(drop=True)
```

```python
tradeDf
```

### 주가지수 데이터

```python
stockIndexDf = pd.DataFrame()
for file_name in glob.glob('data/주가지수*'):
    file = pd.read_excel(file_name)
    months = file.iloc[1].dropna().values
    kospi =file.iloc[2].values[1:]
    kosdaq = file.iloc[3].values[1:]
    for i in range(len(kospi)):
        kospi[i] = kospi[i].replace(',', '')
        
    for i in range(len(kosdaq)):            
        kosdaq[i] = kosdaq[i].replace(',', '')
        
        if kosdaq[i] == '-':
            kosdaq[i] = np.nan
        
    kospi = kospi.astype('float64')    
    kosdaq = kosdaq.astype('float64')
    
    df = pd.DataFrame({'date': months, 'kospi': kospi, 'kosdaq': kosdaq})
    df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y%m월'))
    df['date'] = df['date'].apply(lambda x: x.strftime('%Y-%m'))
    stockIndexDf = pd.concat([stockIndexDf, df]).reset_index(drop=True)
```

```python
stockIndexDf
```

### 수출실적 vs KOSPI

```python
startDt = max(tradeDf['date'].min(), stockIndexDf['date'].min())
endDt = min(tradeDf['date'].max(), stockIndexDf['date'].max())
print(startDt, endDt)
```

```python
kospi = stockIndexDf[(stockIndexDf['date'] >= startDt) & \
                     (stockIndexDf['date'] <= endDt)].set_index('date').loc[:, 'kospi']
trade = tradeDf[(tradeDf['date'] >= startDt) & \
                (tradeDf['date'] <= endDt)].set_index('date').loc[:, 'export [100m$]']
```

```python
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(111)
ln1 = ax1.plot(kospi, label='KOSPI', color='r')
ax2 = ax1.twinx()
ln2 = ax2.plot(trade, label='export', color='b')

ax1.set_xlabel('Date', fontsize=20, labelpad=10)
ax1.set_ylabel('KOSPI', fontsize=20, labelpad=10)
ax2.set_ylabel('Export', fontsize=20, rotation=270, labelpad=20)
ax1.set_xticks(kospi.index[::12])
ax1.tick_params(axis='x', rotation=90)
ax1.tick_params(axis='y', labelsize=15)
ax2.tick_params(axis='y', labelsize=15)

lns = ln1 + ln2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0, fontsize=15)
plt.show()
```

```python
compDf = pd.DataFrame({'kospi': kospi, 'export': trade})
compDf['shift-6'] = compDf['export'].shift(-6)
compDf['shift-5'] = compDf['export'].shift(-5)
compDf['shift-4'] = compDf['export'].shift(-4)
compDf['shift-3'] = compDf['export'].shift(-3)
compDf['shift-2'] = compDf['export'].shift(-2)
compDf['shift-1'] = compDf['export'].shift(-1)
compDf['shift1'] = compDf['export'].shift(1)
compDf['shift2'] = compDf['export'].shift(2)
compDf['shift3'] = compDf['export'].shift(3)
compDf['shift4'] = compDf['export'].shift(4)
compDf['shift5'] = compDf['export'].shift(5)
compDf['shift6'] = compDf['export'].shift(6)
compDf = compDf.dropna()
compDf.corr()['kospi'].sort_values(ascending=False)
```
